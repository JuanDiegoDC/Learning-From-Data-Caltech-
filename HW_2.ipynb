```python
"""
Created on Sat Oct 28 15:37:28 2017

@author: JD
"""
%matplotlib inline
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
np.random.seed(0)
```

```python
"""
Flip 1000 fair-coins, 10 times each independently
and obtain v_1, v_rand, v_min 
"""

def coin_flips(flips=1000):
    coins = np.random.binomial(n=10, p=0.5, size=flips)
    v_all = [np.mean(coin) for coin in coins]
    v_1 = v_all[0]
    v_rand = v_all[np.random.randint(1000)]
    v_min = min(v_all)
    return (v_1, v_min, v_rand)
```

```python
"""
Repeat coin_flips experiment 1000 times and
return the distribution of v_1, v_min, v_rand
as v_1_dist v_min_dist, v_rand_dist
"""

def coin_simulation():
    v_1_dist, v_min_dist, v_rand_dist = [], [], []
    for i in range(1000):
        v_1, v_min, v_rand = coin_flips()
        v_1_dist.append(v_1)
        v_min_dist.append(v_min)
        v_rand_dist.append(v_rand)
    return (v_1_dist, v_min_dist, v_rand_dist)

v_1_dist, v_min_dist, v_rand_dist = coin_simulation()
```

```python
"""
Obtain mean of each respective distribution
"""

nu_1 = np.mean(v_1_dist)
nu_min = np.mean(v_min_dist)
nu_rand = np.mean(v_rand_dist)

print(nu_1)
print(nu_min)
print(nu_rand)
```
